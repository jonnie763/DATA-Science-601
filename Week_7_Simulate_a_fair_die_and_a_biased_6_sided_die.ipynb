```
import numpy as np                      # For random choice 
import pandas as pd                     # Normally this is module I always import 
from matplotlib import pyplot as plt    # For showing differences in data 
```

```
DICE = [1,2,3,4,5,6]                     # Numbers of six sided dice 
PROB = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]    # Probabilties of fair sided dice 

sample_means = []                        # Will append later 
sample_sizes = range(5, 10000, 10)       # Starts at sample size of 5, and increase by 10 until it reaches 10000, user can change this within the code 


sample_means = [np.random.choice(a=DICE, p=PROB, size=s).mean() for s in sample_sizes]  # Finds the mean of considering the probability of a fair dice, so 1/6(1)+ 1/6(2)....1/6(6)= 3.5 

sample_means[-15:]                                                                      # The simulation, with its large sample size, will get a number close to the population mean of 3.5 
```

```
pop_mean = np.mean(DICE) # Theoretical mean  

fig, ax = plt.subplots(figsize=(12,10)) # Graph size 

                                         
ax.plot((0,sample_sizes[-1]), (pop_mean, pop_mean),linewidth=6) #Theoretical mean plotted, to shown convergence towards 3.5 


ax.plot(sample_sizes, sample_means) # From data from previous cell  

                   
ax.set_title("Fair Dice: Coverging to 3.5 Sample Mean", size=16)
ax.set_ylabel("Sample Mean of Fair Dice", size=14)
ax.set_xlabel("Sample Size", size=14)
ax.text(1400, 3.8, r'Biased Dice has 3.75 sample mean, where as fair dice has sample mean of 3.5',fontsize=13) # Annotation to note the differences of two graphs, one dice is fair, the other is biased towards the six thus creating a higher sample mean  
```

```
DICE = [1,2,3,4,5,6]                           # Numbers of six sided dice
PROB = [0.15, 0.15, 0.15, 0.15, 0.15, 0.25]    # Probabilties of biased sided dice 

sample_means_2 = []                            # Will append later 
sample_sizes_2 = range(5, 10000, 10)           # Starts at sample size of 5, and increase by 10 until it reaches 10000, user can change this within the code 


sample_means_2 = [np.random.choice(a=DICE, p=PROB, size=s).mean() for s in sample_sizes_2] # Finds the mean of considering the probability of a fair dice, so .15(1)+ .15(2).....25(6)= 3.75 

sample_means_2[-15:]                                                                       # The simulation, with its large sample size, will get a number close to the population mean of 3.75
```

```
pop_mean = np.mean(sample_means_2)      # Theoretical mean  

fig, ax = plt.subplots(figsize=(12,10)) # Graph size 

                                            
ax.plot((0,sample_sizes_2[-1]), (pop_mean, pop_mean),linewidth=6) #Theoretical mean plotted, to shown convergence towards 3.75 

ax.plot(sample_sizes_2, sample_means_2)                           # From data from previous cell
                   
ax.set_title("Biased Dice: Coverging to 3.75 Sample Mean", size=16) 
ax.set_ylabel("Sample Mean of FBias Dice", size=14)
ax.set_xlabel("Sample Size", size=14)

ax.text(1400, 4.0, r'Biased Dice has 3.75 sample mean, where as fair dice has sample mean of 3.5',fontsize=13) # Annotation to note the differences of two graphs, one dice is fair, the other is biased towards the six thus creating a higher sample mean  
```
