```python
# Citation https://machinelearningmastery.com/clean-text-machine-learning-python/
# load text
filename = '52256-0.txt'                    # Combine code into one chunk, this removes the punctuation 
file = open(filename, encoding="utf8")

text = file.read()
file.close()
# split into words by white space
words = text.split()
# remove punctuation from each word

import string
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in words]
print(stripped[:100])
```

```python
# using list comprehension
listToStr = ' '.join([str(elem) for elem in stripped])
  
print(listToStr[:100]) 
len(listToStr)            # I need to turn the list to a string, to use a library that removes the stop words 
```

```python
import gensim
from gensim.parsing.preprocessing import remove_stopwords

listToStr
filtered_sentence_1 = remove_stopwords(listToStr) # I need to turn the list to a string, to use a library that removes the stop words in the next line of code 


print(filtered_sentence_1[:100]) # Print the file with removed stop words and punctuation 

```
